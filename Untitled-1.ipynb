```python

# ==========================================
# 6. ÂìÅË©ûÂèñÂæóÔºàÂÖÉ self.get_posÔºâ
# ==========================================
@lru_cache(maxsize=4096)
def get_pos(word):
    if not word:
        return ""
    try:
        tagger = MeCab.Tagger("")
        parsed = tagger.parse(word)
        if not parsed:
            return ""
        parts = parsed.splitlines()[0].split("\t")
        features = parts[1].split(",")
        return features[0]
    except:
        return ""
```

```python
# ==========================================
# 8. Ê©üËÉΩÔºöÂàÜ„Åã„Å°Êõ∏„Åç
# ==========================================
def tokenize_text():
    global original_text, tokens, word_freq, pos_cache

    text = text_area.get("1.0", tk.END).strip()
    if not text:
        messagebox.showwarning("Ë≠¶Âëä", "„ÉÜ„Ç≠„Çπ„Éà„ÇíÂÖ•Âäõ„Åó„Å¶„Åè„Å†„Åï„ÅÑ")
        return

    original_text = text

    wakati = mecab.parse(text)
    wakati = re.sub(r'[\u200B\u200C\u200D\uFEFF]', '', wakati)
    wakati = wakati.replace("\u3000", " ")
    words = wakati.split()

    tokens = [w for w in words if w not in stop_words and len(w) > 1]
    pos_cache = [get_pos(w) for w in tokens]
    word_freq = Counter(tokens)

    edit_area.delete("1.0", tk.END)
    edit_area.insert("1.0", " ".join(tokens))

    refresh_word_list()
    notebook.select(1)
```

```python

# ==========================================
# 9. ÂçòË™û„É™„Çπ„ÉàÊõ¥Êñ∞
# ==========================================
def refresh_word_list():
    word_listbox.delete(0, tk.END)
    for word, count in word_freq.most_common():
        word_listbox.insert(tk.END, f"{word} ({count})")
```

```python

# ==========================================
# 10. ÂèØË¶ñÂåñÔºàWordCloud / Network / FreqÔºâ
# ==========================================
def visualize():
    text = edit_area.get("1.0", tk.END).strip()
    if not text:
        messagebox.showwarning("Ë≠¶Âëä", "ÂçòË™û„Éá„Éº„Çø„Åå„ÅÇ„Çä„Åæ„Åõ„Çì")
        return

    freq = Counter(text.split())
    if not freq:
        return

    generate_wordcloud(freq)
    generate_network(text.split(), freq)
    generate_freq(freq)

    notebook.select(2)
```

```python
# ==========================================
# 11. WordCloud
# ==========================================
def generate_wordcloud(freq):
    for w in wordcloud_frame.winfo_children():
        w.destroy()

    wc = WordCloud(
        width=1000, height=600,
        background_color="white",
        font_path=font_path
    ).generate_from_frequencies(freq)

    fig, ax = plt.subplots(figsize=(12,7))
    ax.imshow(wc); ax.axis("off")

    canvas = FigureCanvasTkAgg(fig, wordcloud_frame)
    canvas.get_tk_widget().pack(fill="both", expand=True)
    canvas.draw()

```

```python

# ==========================================
# 12. ÂÖ±Ëµ∑„Éç„ÉÉ„Éà„ÉØ„Éº„ÇØ
# ==========================================
def generate_network(tokens, freq):
    for w in network_frame.winfo_children():
        w.destroy()

    G = nx.Graph()

    for i in range(len(tokens)-1):
        G.add_edge(tokens[i], tokens[i+1])

    fig, ax = plt.subplots(figsize=(12,8))
    pos = nx.spring_layout(G, k=1.5)

    nx.draw(G, pos, with_labels=True,
            font_family=font_name, node_size=300,
            ax=ax)

    canvas = FigureCanvasTkAgg(fig, network_frame)
    canvas.get_tk_widget().pack(fill="both", expand=True)
    canvas.draw()

```

```python

# ==========================================
# 13. È†ªÂ∫¶„Ç∞„É©„Éï
# ==========================================
def generate_freq(freq):
    for w in freq_frame.winfo_children():
        w.destroy()

    top = dict(freq.most_common(30))

    fig, ax = plt.subplots(figsize=(12,8))
    ax.barh(list(top.keys()), list(top.values()))
    ax.invert_yaxis()

    canvas = FigureCanvasTkAgg(fig, freq_frame)
    canvas.get_tk_widget().pack(fill="both", expand=True)
    canvas.draw()
```
